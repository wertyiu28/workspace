### 문제 5. 빈 데이터 판별기(심화)

사용자로부터 입력받은 검색어 `search_keyword`가 비어있는지 확인하는 코드를 작성하세요.

- 검색어가 있으면: **" '검색어'를 검색합니다."** 출력
- 검색어가 비어있거나 공백만 있다면(strip 활용): **"검색어를 입력해주세요."** 출력

```python
search_keyword = " 안녕~~"      # 예시: 공백만 입력된 경우

clean_keyword = search_keyword.strip()

if search_keyword : 
    print(f"'{clean_keyword}'를 검색합니다.")

else : print("검색어를 입력해주세요.")

```
