### 문제 2. 리스트의 중복 제거와 정렬

다음 리스트에서 중복된 과일 이름을 제거하고, 가나다 순(오름차순)으로 정렬된 리스트를 출력하세요.

fruits = ["사과", "바나나", "사과", "포도", "바나나", "체리"]

```python
fruits = ["사과", "바나나", "사과", "포도", "바나나", "체리"]

fruits_list = set(fruits)
print(f'중복된 과일 목록 제거: {fruits_list}')

sorted_fruits = sorted(fruits_list)
print(f'정렬된 과일 목록 : {sorted_fruits}')
```
